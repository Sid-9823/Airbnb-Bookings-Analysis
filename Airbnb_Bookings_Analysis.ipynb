# **Exploratory Data Analysis and Visualization**

**What is EDA ?**

***Exploratory Data Analysis can be considered as a process which is used to visualise data and it helps to uncover the underlying pattern and relations between the dependent and independent variable's which can be useful in decision making.***

##**What is the number of Airbnb apartments in each neighbourhood group ?**

```python
# Plotting a barplot.

plt.figure(figsize=(10, 5))
ax = df.groupby('ngrp')['id'].count().plot(kind='bar', color='skyblue', edgecolor='black')

# To annotate the countplot.

for p in ax.patches:
  ax.annotate(f"{p.get_height():.2f}", (p.get_x() + p.get_width() / 2, p.get_height()),
  ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')

plt.xlabel('Neighbourhood Groups', fontsize=14)
plt.title('Number of Airbnb apartments in different neighbourhood groups', fontsize=16, fontweight = 'bold')
plt.xticks(rotation=0, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
```

* ***Manhattan has the highest number of Airbnb apartments, followed by Brooklyn.***

##**How much does a unit cost on average in each neighbourhood group ?**

```python
# Create the figure and axis objects

plt.figure(figsize=(10, 5))
ax = df.groupby('ngrp')['price'].mean().plot(kind='bar', color='skyblue', edgecolor='black')

# To annotate the countplot.

for p in ax.patches:
  ax.annotate(f"{p.get_height():.2f}", (p.get_x() + p.get_width() / 2, p.get_height()),
  ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')

plt.xlabel('Neighbourhood Groups', fontsize=14)
plt.ylabel('Average Price', fontsize=14)
plt.title('Unit average cost in each neighbourhood group', fontsize=16, fontweight = 'bold')
plt.xticks(rotation=0, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
```

* ***Unit costs are highest in Manhattan, followed by Brooklyn and Staten Island.***

* ***Lowest unit cost is in Bronx.***

* ***Because all of NYC's tourist attractions are concentrated in Manhattan and Brooklyn, travellers prefer to stay in these areas, allowing owners to charge greater costs as a result.***

* ***Airbnb should consider investing in any of these two neighbourhood organisations.***

##**What are the top ten most affordable neighbourhoods ?**

```python
# Creating a dataframe using groupby.

c = pd.DataFrame(df.groupby('neigh')['price'].mean().sort_values(ascending = True).head(10).reset_index())
c
```

* ***Filtering out the top 10 cheap neighbourhoods using group and mean in a dataframe.***

```python
# Plotting a barplot using the above created dataframe.

plt.figure(figsize = (15,6))
ax = sns.barplot(x = 'neigh', y = 'price', data = c, palette = 'viridis')

# To annotate the barplot.

for p in ax.patches:
  ax.annotate(f"{p.get_height():.2f}", (p.get_x() + p.get_width() / 2, p.get_height()),
  ha='center', va='bottom', fontsize=12, fontweight = 'bold',color='black')


plt.title('Top - ten most affordable neighbourhoods', fontsize = 16, fontweight = 'bold')
plt.xlabel('Neighbourhood', fontsize=14)
plt.ylabel('Average Price', fontsize=14)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
```

* ***Among the top 10 cheap neighbourhood Airbnb units, Bull's Head, Hunt's Point, and Tremont units are the cheapest.***

##**What is the bare minimum of nights that hosts prefer in all neighbourhood groups ?**

```python
# Creating a dataframe using groupby.

n = pd.DataFrame(df.groupby('ngrp')['minn'].mean().reset_index())
n
```

* ***Grouping neighbourhood groups using groupby with the mean of the minimum number of night.***

```python
# Plotting a barplot using the above created dataframe.

plt.figure(figsize = (15,5))
ax = sns.barplot(x = 'ngrp', y = 'minn', palette = 'viridis', data = n)

# To annotate the barplot.

for p in ax.patches:
  ax.annotate(f"{p.get_height():.2f}", (p.get_x() + p.get_width() / 2, p.get_height()),
  ha='center', va='bottom', fontsize=12, fontweight = 'bold',color='black')


plt.title('Minimum number of nights that hosts prefer considering neighbourhood groups', fontsize = 16, fontweight = 'bold')
plt.xlabel('Neighbourhood Groups', fontsize=14)
plt.ylabel('Minimum number of nights', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
```

* ***In Manhattan, hosts prefer a minimum stay of around 8 days.***

* ***In Brooklyn, hosts prefer a minimum stay of around 6 days.***

* ***In Queens, hosts prefer a minimum stay of around 5 days.***

## **Correlation Heatmap**

**What is a correlation heatmap?**

* ***We can define a heatmap as a visual representation of the correlation between all the variables present in the data frame.***

* ***To represent this, it uses a system of colour palettes ranging from light to dark, demonstrating low to high positive or negative correlation.***

```python
# Correlation heatmap of our dataset.

plt.figure(figsize = (18,8) )
sns.heatmap(df.corr(), annot = True, cmap = 'YlGnBu')
```

* ***Host ID and listing ID are showing a correlation of 59%, which is obvious.***

* ***23% correlation between calculated host listing count and availability.***

* ***17% correlation between the number of reviews and availability.***

* ***16% correlation between the number of minimum nights to stay and availability.***

* ***20% correlation between host ID and availability.***
